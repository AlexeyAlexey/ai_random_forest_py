We load trained Rrandom Forest model generated by [ai_segments_without_pca.ipynb](https://github.com/AlexeyAlexey/ai_random_forest_py/blob/main/ai_segments_without_pca.ipynb)

I use the same dataset that we use to train a model. I do it only to test that model is loaded

```python
from sklearn.ensemble import RandomForestClassifier

import matplotlib.pyplot as plt

# statistical data visualization
import seaborn as sns

# Loading pandas
import pandas as pd

# Loading numpy
import numpy as np

# is used to load saved Random Forest
# https://joblib.readthedocs.io/en/latest/
import joblib
```

```python
# load, no need to initialize the loaded_rf
loaded_clf = joblib.load("./random_forest.joblib")
```

```python
target = "paying" # It is what we try to predict

# a file was generated for 'session_29d'
df = pd.read_csv("/vagrant/ai_random_forest_py/contacts/202402070946/segments.csv", low_memory=False) # returns DataFrame

df.head()
```

```python
del df["created_date"]
del df["id"]

df.head()
```

```python
print('Paying', df[df['paying'] == 1].shape[0])
print('Not Paying', df[df['paying'] == 0].shape[0])
```

```python
# Create a list of the feature column's names


# print(df.shape[1] - 2)
features = df.columns[:(df.shape[1] - 1)]


# View features
print(f'Features: {features}')

df.head()
```

```python
x = df[features]
y = df['paying']
```

```python
y_pred = loaded_clf.predict(x)

y_pred.astype(bool)
```

```python
from sklearn.metrics import recall_score

display(pd.crosstab(y.astype(bool), y_pred.astype(bool), rownames = ['Actual Paying'], colnames = ['Predicted Paying']))

display('Baseline Random Forest recall score', recall_score(y, y_pred))
```
